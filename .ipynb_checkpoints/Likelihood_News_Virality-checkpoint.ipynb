## Converting extracted News articles to conform to the Training set taken

```python
from nltk.tokenize import word_tokenize 
from nltk.corpus import stopwords
from textblob import TextBlob
import datefinder
import datetime  
from datetime import date

'''
Stopwords, negative and positive words in english 
'''
stopwords=set(stopwords.words('english'))
pos_words=[]
neg_words=[]

def tokenize(text):
    '''
    Function to tokenize the text
    '''
    text=text
    return word_tokenize(text)

def abs(x): 
    if (x < 0): 
        return (x*(-1))
    else: 
        return (x)

def rate_unique(words):
    '''
    Function to tokenize the text
    '''
    words=tokenize(words)
    no_order = list(set(words))
    rate_unique=len(no_order)/len(words)
    return rate_unique

def rate_nonstop(words):
    words=tokenize(words)
    filtered_sentence = [w for w in words if not w in stopwords]
    rate_nonstop=len(filtered_sentence)/len(words)
    no_order = list(set(filtered_sentence))
    rate_unique_nonstop=len(no_order)/len(words)
    return rate_nonstop,rate_unique_nonstop

def avg_token(words):
    words=tokenize(words)
    length=[]
    for i in words:
        length.append(len(i))
    return np.average(length)


def day(article_text):
    article=article_text
    if len(list(datefinder.find_dates(article)))>0:
        date=str(list(datefinder.find_dates(article))[0])
        date=date.split()
        date=date[0]
        year, month, day = date.split('-')     
        day_name = datetime.date(int(year), int(month), int(day)) 
        return day_name.strftime("%A")
    return "Tuesday"

def polar(words):
    all_tokens=tokenize(words)
    for i in all_tokens:
        analysis=TextBlob(i)
        polarity=analysis.sentiment.polarity
        if polarity>0:
            pos_words.append(i)
        if polarity<0:
            neg_words.append(i)
    return pos_words,neg_words

def rates(words):
    words=polar(words)
    pos=words[0]
    neg=words[1]
    all_words=words
    global_rate_positive_words=(len(pos)/len(all_words))/100
    global_rate_negative_words=(len(neg)/len(all_words))/100
    pol_pos=[]
    pol_neg=[]
    for i in pos:
        analysis=TextBlob(i)
        pol_pos.append(analysis.sentiment.polarity)
        avg_positive_polarity=analysis.sentiment.polarity
    for j in neg:
        analysis2=TextBlob(j)
        pol_neg.append(analysis2.sentiment.polarity)
        avg_negative_polarity=analysis2.sentiment.polarity
    min_positive_polarity=min(pol_pos)
    max_positive_polarity=max(pol_pos)
    min_negative_polarity=min(pol_neg)
    max_negative_polarity=max(pol_neg)
    avg_positive_polarity=np.average(pol_pos)
    avg_negative_polarity=np.average(pol_neg)
    
    rate_positive_words = (len(pos) / (len(pos) + len(neg)))
    rate_negative_words = (len(neg) / (len(pos) + len(neg)))
    
    return global_rate_positive_words,global_rate_negative_words,avg_positive_polarity,min_positive_polarity,max_positive_polarity,avg_negative_polarity,min_negative_polarity,max_negative_polarity,rate_positive_words,rate_negative_words
```

```python
newsData1=[]

for i in news:
    
    info={}
    
    article = Article(i, language="en") # en for English 
    article.download() 
    article.parse()
    analysis=TextBlob(article.text)
    
    polarity=analysis.sentiment.polarity
    
    title_analysis=TextBlob(article.title)
    
    # Started filling the data
    info['text']=article.text
    info['n_tokens_title']=len(tokenize(article.title))
    info['n_tokens_content']=len(tokenize(article.text))
    info['n_unique_tokens']=rate_unique(article.text)
    info['n_non_stop_words']=rate_nonstop(article.text)[0]
    info['n_non_stop_unique_tokens']=rate_nonstop(article.text)[1]
    info['num_hrefs']=article.html.count("https://timesofindia.indiatimes.com")
    info['num_imgs']=len(article.images)
    info['num_videos']=len(article.movies)
    info['average_token_length']=avg_token(article.text)
    info['num_keywords']=len(article.keywords)
    
    if "life-style" in article.url:
        info['data_channel_is_lifestyle']=1
    else:
        info['data_channel_is_lifestyle']=0
        
    if "etimes" in article.url:
        info['data_channel_is_entertainment']=1
    else:
        info['data_channel_is_entertainment']=0
        
    if "business" in article.url:
        info['data_channel_is_bus']=1
    else:
        info['data_channel_is_bus']=0
        
    if "social media" or "facebook" or "whatsapp" in article.text.lower():
        data_channel_is_socmed=1
        data_channel_is_tech=0
        data_channel_is_world=0
    else:
        data_channel_is_socmed=0
        
    if ("technology" or "tech" in article.text.lower()) or ("technology" or "tech" in article.url):
        data_channel_is_tech=1
        data_channel_is_socmed=0
        data_channel_is_world=0
    else:
        data_channel_is_tech=0
        
    if "world" in article.url:
        data_channel_is_world=1
        data_channel_is_tech=0
        data_channel_is_socmed=0
    else:
        data_channel_is_world=0
        
    info['data_channel_is_socmed']=data_channel_is_socmed
    info['data_channel_is_tech']=data_channel_is_tech
    info['data_channel_is_world']=data_channel_is_world
    
    if day(i)=="Monday":
        info['weekday_is_monday']=1
    else:
        info['weekday_is_monday']=0
        
    if day(i)=="Tuesday":
        info['weekday_is_tuesday']=1
    else:
        info['weekday_is_tuesday']=0
        
    if day(i)=="Wednesday":
        info['weekday_is_wednesday']=1
    else:
        info['weekday_is_wednesday']=0
        
    if day(i)=="Thursday":
        info['weekday_is_thursday']=1
    else:
        info['weekday_is_thursday']=0
        
    if day(i)=="Friday":
        info['weekday_is_friday']=1
    else:
        info['weekday_is_friday']=0
        
    if day(i)=="Saturday":
        info['weekday_is_saturday']=1
        info['is_weekend']=1
    else:
        info['weekday_is_saturday']=0
        
    if day(i)=="Sunday":
        info['weekday_is_sunday']=1
        info['is_weekend']=1
    else:
        info['weekday_is_sunday']=0
        info['is_weekend']=0
        
    info['global_subjectivity']=analysis.sentiment.subjectivity
    info['global_sentiment_polarity']=analysis.sentiment.polarity
    
    info['global_rate_positive_words']=rates(article.text)[0]
    info['global_rate_negative_words']=rates(article.text)[1]
    info['avg_positive_polarity']=rates(article.text)[2]
    info['min_positive_polarity']=rates(article.text)[3]
    info['max_positive_polarity']=rates(article.text)[4]
    info['avg_negative_polarity']=rates(article.text)[5]
    info['min_negative_polarity']=rates(article.text)[6]
    info['max_negative_polarity']=rates(article.text)[7]
    info['rate_positive_words']=rates(article.text)[8]
    info['rate_negative_words']=rates(article.text)[9] 
    
    info['title_subjectivity']=title_analysis.sentiment.subjectivity
    info['title_sentiment_polarity']=title_analysis.sentiment.polarity
    
    info['abs_title_subjectivity'] = abs(title_analysis.sentiment.subjectivity)
    info['abs_title_sentiment_polarity']=abs(title_analysis.sentiment.polarity)

    newsData1.append(info)
```

### Data Preprocessing and Cleaning

```python
def clean_cols(data):
    '''Clean the column names'''
    clean_col_map = {x: x.lower().strip() for x in list(data)}
    return data.rename(index=str, columns=clean_col_map)
```

Removing the followig fields
- URL (as its not important)
- timedelta (as it only has news of today so all test values for this parameter will be 0)
- LDA00, LDA01, LDA02, LDA03, LDA04 (as LDA topics can't be described without proper knowledge) 
- Num_self_refs (as there are almost no self reference in the articles that have been scrapped)
- kw_min_min, kw_max_min, kw_avg_min, kw_min_max, kw_max_max, kw_avg_max, kw_min_avg, kw_max_avg, kw_avg_avg (as keyword cannot be descirbed as Worst, Average or Best)
- self_reference_min_shares, self_reference_max_shares, self_reference_avg_sharess (as news articles have no self references)

```python
def reject_outliers(shares):
    '''
    function to remove outliers, which leaves out the values (here shares) is larger than its mean +/- 3 SD
    '''
    u = np.median(dataframe['shares'])
    s = np.std(dataframe['shares'])
    filtered= [e for e in (dataframe['shares']) if (u - 2 * s < e < u + 2 * s)]
    return filtered
```

### Hyperparameters and Feature Importance

##### Grid Search to find optimal hyperparameters

```python
param_grid = [
     {'n_estimators': [3,7,10], 'max_features':[2,3,5,7,10]},
    {'bootstrap': [False], 'n_estimators': [3,7,10],'max_features': [2,3,5,7,10]}
]

forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                          scoring="neg_mean_squared_error")

grid_search.fit(X_train_set, y_train_set)
```

```python
cvres=grid_search.cv_results_
for mean_score, params in zip(cvres['mean_test_score'], cvres['params']):
    print(np.sqrt(-mean_score), params)
```

```python
print(grid_search.best_params_)
```

##### Finding the optimal number of hyperparameters through Randomized Search


The Grid Search is fine when we have only relatively few combinations. But when the hyperparameter search space is large, like in this case, RandomizedSearchCV is preferable. Instead of trying out all possible combinations, it evaluates a given number of random combinations by selecting a random value for each hyperparameter at every iteration.

The limits for the n_estimators and max_features are set to 20 and 30, respectively, to keep things from getting too complicated.

```python
param_distribs = {
        'n_estimators': randint(low=1, high=20),
        'max_features': randint(low=1, high=25),
    }

forest_reg = RandomForestRegressor(random_state=42)
rnd_search = RandomizedSearchCV(forest_reg, param_distributions=param_distribs,
                                n_iter=30, cv=5, scoring='neg_mean_squared_error', random_state=42)

rnd_search.fit(X_train_set, y_train_set)
```

```python
cvres = rnd_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

We seem to obtain the best solution - i.e. with lowest RMSE - by setting max_features hyperparameter to 5 and the n_estimators hyperparameter to 19; RMSE: 3216.124198360326
